```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Float, Text

# Sample data for demonstration purposes


# Create DataFrame
df = pd.read_csv("assistant/data/courssistant_main.csv")

# Add an id column
df['id'] = range(1, len(df) + 1)

# Reorder columns if necessary
df = df[['id'] + [col for col in df.columns if col != 'id']]

# Establish a connection to the MySQL database
DATABASE_TYPE = 'mysql+pymysql'
DBAPI = 'pymysql'
HOST = 'localhost'
USER = 'root'
PASSWORD = ''
DATABASE = 'duck_demo'

DATABASE_URL = f"{DATABASE_TYPE}://{USER}:{PASSWORD}@{HOST}/{DATABASE}"
engine = create_engine(DATABASE_URL)
# Course Name,University,Difficulty Level,
# Course Rating,Course URL,Course Description,
# Skills,Specialized,Name Adjusted,Description Adjusted,
# Sequence,Vector,IDF,BERT_Encoded,Course_Name_Encoded

# Define the metadata and table structure
metadata = MetaData()
courses = Table('courses', metadata,
                Column('id', Integer, primary_key=True),
                Column('CourseName', String(255)),
                Column('University', String(255)),
                Column('DifficultyLevel', String(50)),
                Column('CourseRating', Float),
                Column('CourseDescription', Text),
                Column('Skills', Text),
                Column('Specialized', String(255)),
                Column('SkillVectors', Text)
                )

# Create the table
metadata.create_all(engine)

# Insert data into the SQL table
df.to_sql('courses', engine, if_exists='append', index=False)

print("Data imported successfully.")
```
